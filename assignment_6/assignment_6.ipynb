```python
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm,y):
    df_cm = pd.DataFrame(cm, index = [i+1 for i in np.unique(y)],
                  columns = [i+1 for i in np.unique(y)])
    plt.figure()
    sn.heatmap(df_cm, annot=True)
    plt.title('Confusion matrix')
    plt.xlabel('Predicted class')
    plt.ylabel('Actual class')
    plt.show()
```

```python
import scipy
from scipy import io
from sklearn import model_selection 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from toolbox.classification_plot import classification_plot

def plot_data_w_attributes(X_train_, y_train_, x_test_, y_test_, distance_method, KNN_Value, i):
    if distance_method == "seuclidean":
        variances = np.var(X_train_, axis=0)
        KNC = KNeighborsClassifier(n_neighbors=KNN_Value,metric=distance_method, metric_params={'V':variances})
    else:
        KNC = KNeighborsClassifier(n_neighbors=KNN_Value, metric=distance_method)
    KNC.fit(X_train_, y_train_)
    prediction = KNC.predict(x_test_)
    classification_plot(x_test_, y_test_, prediction, X_train_, y_train_)
    plt.show()
    print("Figure " + str(i) + ": This figure shows synth" + str(i) + " data using K Nearest Neighbours (K = " + str(KNN_Value) + ") using the distance metric " + distance_method)

data = []
Xtrain = []
ytrain = []
Xtest = []
ytest = []

for x in  range(4):
    data.append(scipy.io.loadmat(f"./data/synth{x+1}.mat"))
    Xtrain.append(data[x]['X_train'])
    ytrain.append(data[x]['y_train'].ravel())
    Xtest.append(data[x]['X_test'])
    ytest.append(data[x]['y_test'].ravel())
    

for i in range(4):
    KNC = KNeighborsClassifier(n_neighbors=3)
    KNC.fit(Xtrain[i], ytrain[i])
    prediction = KNC.predict(Xtest[i])
    classification_plot(Xtest[i], ytest[i], prediction,Xtrain[i], ytrain[i])
    plt.show()
    print("Figure " + str(i+1) + ": This figure shows synth" + str(i+1) + " data using K Nearest Neighbours (K = 3)")

for i in range(4):
    KNC = KNeighborsClassifier(n_neighbors=45)
    KNC.fit(Xtrain[i], ytrain[i])
    prediction = KNC.predict(Xtest[i])
    classification_plot(Xtest[i], ytest[i], prediction,Xtrain[i], ytrain[i])
    plt.show()
    print("Figure " + str(i+5) + ": This figure shows synth" + str(i+1) + " data using K Nearest Neighbours (K = 45)")


best_distance = ['euclidean','cosine','seuclidean','euclidean']
best_k = [2,3,1,15]

k=9
for i in range(4):
    plot_data_w_attributes(Xtrain[i], ytrain[i], Xtest[i], ytest[i], best_distance[i], best_k[i], k)
    k = k + 1
```
